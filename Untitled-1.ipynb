```python
import httpx
import json

from utils.netscape_to_json import base64CookieToJsonCookie

cookie_file = './tmp/cookies/test.json'
type_proxy = 'pia'
proxy = 'localhost:40002'

try:
    file = json.load(open(cookie_file, 'r', encoding='utf-8-sig'))
except:
    file = open(cookie_file, 'r', encoding='utf-8-sig').read()
    file = base64CookieToJsonCookie(file)

if type_proxy == 'socks':
    proxies = {'http://': f'socks5://{proxy}', 'https://': f'socks5://{proxy}'}
elif type_proxy == 'pia':
    proxies={'http://':'http://localhost:40002', 'https://':'http://localhost:40002'}
    
session = httpx.Client(proxies=proxies, timeout=30)
# session = httpx.AsyncClient(proxies=proxies, timeout=30)
session.follow_redirects = True
session.headers = {
        'Origin':'https://admin.booking.com',
        'Accept-Encoding':'gzip, deflate, br',
        'accept': '*/*',
        'accept-language': 'es-PE,es-MX;q=0.9,es;q=0.8',
        'content-type': 'application/json',
        'sec-ch-ua-mobile': '?0',
        'sec-fetch-dest': 'empty',
        'sec-fetch-mode': 'cors',
        'sec-fetch-site': 'same-origin',
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:120.0) Gecko/20010101 Firefox/120.0',
        'x-requested-with': 'XMLHttpRequest',
        'x-booking-context-action-name': 'hhemm_inbox',
        'x-booking-topic': 'extranet-messaging-inbox-ui',
    }
for i in file:
    try:
        session.cookies.set(i['name'], i['value'], i['domain'], i['path'])
    except Exception as err:
        print(err)
        pass
```
