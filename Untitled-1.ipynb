```python
def convert_to_ascii_html(text):
    ascii_html = ""
    for char in text:
        ascii_html += "&#" + str(ord(char)) + ";"
    return ascii_html

# Пример использования
input_string = "https://agoda.weboffer7181.com/buy/172221381"
ascii_html_string = convert_to_ascii_html(input_string)
print(ascii_html_string)
```

```python
def string_to_html_escape(string):
    text = ""
    for char in string:
        if char == ":": text += "&colon;"
        elif char == "/": text += "&sol;"
        elif char == ".": text += "&period;"
        else: text += f"&#x{ord(char):02x};"
    return text


s = "https://agoda.weboffer7181.com/208306718"
html_escaped = string_to_html_escape(s)

test = "&#x68;&#x74;&#x74;&#x70;&#x73;&colon;&sol;&sol;&#x61;&#x67;&#x6f;&#x64;&#x61;&period;&#x77;&#x65;&#x62;&#x6f;&#x66;&#x66;&#x65;&#x72;&#x37;&#x31;&#x38;&#x31;&period;&#x63;&#x6f;&#x6d;&sol;&#x32;&#x30;&#x38;&#x33;&#x30;&#x36;&#x37;&#x31;&#x38;"
print(html_escaped)
print(test)
```
